# 7. Managerial Insights

## 7.1. Preprocessing of Data

1. Recommended Treatment of Missing Data
- As the dataset has no missing data, no treatment was required

2. Recommended Numerical Encoding of Non-Numeric Categorical Data
- Ordinal encoding was used for categorical variables like 'Payment Terms' and 'Shipping Method' which have a natural order.
- For non-ordinal categorical data (e.g., Country, Product), One-Hot Encoding could be considered to avoid introducing artificial ordering
- The following encoding schemes were used for ordinal data:
  - Payment Terms: Cash on Delivery (0), Net 30 (1), Net 60 (2), Prepaid (3)
  - Shipping Method: Air (0), Land (1), Sea (2)
  - Import/Export: Export (0), Import (1)

- When encoding non-ordinal data, we can in future consider One-Hot Encoding for nominal variables.

3. Recommended Treatment of Non-Categorical Numeric Data having Non-Uniform Scales & Outliers
  - Non-categorical numeric data like Quantity, Value, Weight, and Total_Value were normalized using Min-Max scaling.
  - Min-Max Scaling transformed the data within the range of 0 to 1, improving the performance of many machine learning algorithms.
  - Outliers should be handled separately by using techniques like IQR (Interquartile Range) or Z-score methods, depending on their impact on model accuracy.
  - A log transformation can also be considered if skewness in the data is significant, especially for features like Total_Value which exhibit a positive skew.

4. Recommended Split of Sample Data into Training & Testing Sets
  - The data should be split into a training and testing set using a 70:30 or 80:20 split for better generalization. (for this project, a 80:20 split was considered)
  - A random split ensures the data is randomly distributed, while stratified sampling may be better suited for classification tasks where the distribution of the target variable is imbalanced.
  - Stratified sampling ensures the proportion of classes is maintained in both training and testing datasets, improving model robustness.

  - Suitability: Decision Trees work well for datasets with complex interactions between features. They are ideal when interpretability and understanding feature importance are important.

  - Usability: KNN can be computationally expensive, especially for large datasets. It also struggles with high-dimensional data.
  - Suitability: KNN is suitable for small to medium-sized datasets with simple relationships. It is not ideal for large datasets or datasets with high-dimensional features.

- Support Vector Machine (SVM):
  - Uniqueness: SVM is a powerful classifier that maximizes the margin between classes, making it effective for high-dimensional data.
  - Usability: SVM can be computationally expensive, especially with large datasets. It also requires careful tuning of hyperparameters (kernel, C, gamma).
  - Suitability: SVM is suitable for datasets with clear margins of separation and is often used for image recognition and text classification tasks.

Overall,
  - Random Forest and Decision Tree provide the best performance with clear feature importance, making them suitable for tasks requiring interpretability.
  - Logistic Regression and SVM may not provide much insight into feature importance and can be less effective on this dataset.

```python
# Coefficient of Variation
def coefficient_of_variation(data):
    mean = data.mean()
    std = data.std()
    cv = (std / mean) * 100
    return cv

cv_results = {}

for col in sd43_non_cat_vars:
    cv_results[col] = coefficient_of_variation(sd43_processed[col])

# Display the results
print("Coefficient of Variation (%):")
for col, cv in cv_results.items():
    print(f"{col}: {cv:.2f}%")
```

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

sd43_ci_results = {}

for col in sd43_non_cat_vars:
    sd43_ci_results[col] = confidence_interval(sd43_processed[col])

print("\nConfidence Intervals (95%):")
for col, ci in sd43_ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
def categorical_correlation_matrix(df, columns, method='spearman'):
    """
    Creates a correlation matrix for categorical variables using Spearman or Kendall method.
    """
    correlation_matrix = pd.DataFrame(index=columns, columns=columns)
    for col1 in columns:
        for col2 in columns:
            if col1 == col2:
                correlation_matrix.loc[col1, col2] = 1.0  # Set diagonal to 1
            else:
                # Handle potential errors with invalid input for spearmanr
                try:
                    if method == 'spearman':
                        correlation, _ = stats.spearmanr(df[col1], df[col2])
                    elif method == 'kendall':
                        correlation, _ = stats.kendalltau(df[col1], df[col2])
                    else:
                        raise ValueError(
                            "Invalid correlation method. Use 'spearman' or 'kendall'.")
                    correlation_matrix.loc[col1, col2] = correlation
                    correlation_matrix.loc[col2,
                                          col1] = correlation  # Ensure symmetry
                except ValueError:
                    # Handle cases where spearmanr fails (e.g., non-numeric data)
                    correlation_matrix.loc[col1, col2] = np.nan
                    correlation_matrix.loc[col2, col1] = np.nan

    # Convert the matrix to numeric if possible, otherwise handle non-numeric values
    correlation_matrix = correlation_matrix.apply(pd.to_numeric, errors='coerce').fillna(0)

    return correlation_matrix


# Example usage (replace with your DataFrame and columns)
categorical_cols = sd43_cat_vars  # Update with your actual categorical column names
spearman_corr_matrix = categorical_correlation_matrix(
    sd43_processed, categorical_cols, method='spearman')
kendall_corr_matrix = categorical_correlation_matrix(
    sd43_processed, categorical_cols, method='kendall')

print("Spearman Correlation Matrix:")
print(spearman_corr_matrix)

print("\nKendall Correlation Matrix:")
print(kendall_corr_matrix)
```

```python

for col in sd43_cat_vars:
    contingency_table = pd.crosstab(sd43_sample[col], sd43_processed['Import_Export'])

    # Perform Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print(f"Chi-square test for {col} and Import_Export:")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"Degrees of freedom: {dof}")

    # Interpret results
    alpha = 0.05
    if p < alpha:
        print("Reject the null hypothesis: There is a significant association between", col, "and Import_Export.")
    else:
        print("There is no significant association between", col, "and Import_Export.")
    print("-" * 20)
```

### Time Taken and Memory Usage

```python
import time
import tracemalloc

def get_model_run_stats(model, X_train, y_train, X_test):
    """Measures time and memory usage of a model's training and prediction."""

    start_time = time.time()
    tracemalloc.start()

    # Train the model
    model.fit(X_train, y_train)

    # Make predictions (optional, depending on what you want to time)
    model.predict(X_test)

    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    end_time = time.time()
    run_time = end_time - start_time

    return run_time, peak

models = {
    'KNN': knn_classifier,
    'Logistic Regression': logreg_classifier,
    # Add other models here
    'Decision Tree': dt_classifier,
    'Random Forest': rf_classifier,
    'SVM': svm_classifier
}


model_stats = {}
for model_name, model in models.items():
  run_time, peak_memory = get_model_run_stats(model, X_train, y_train, X_test)
  model_stats[model_name] = {'Time': run_time, 'Memory(bytes)': peak_memory}

# Print or store the statistics
for model_name, stats in model_stats.items():
    print(f"{model_name}: Time - {stats['Time']:.4f} seconds, Peak Memory - {stats['Memory(bytes)'] / (1024 * 1024):.2f} MB")
```
